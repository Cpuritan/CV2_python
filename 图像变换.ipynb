### HSV格式进行灰度拉伸

```python
#HSV增加明度
def HSV_bright(x):
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            if x[i][j][2]*1.1>100:
                x[i][j][2]=100
            else:
                x[i][j][2]=x[i][j][2]*1.1
    return x

#HSV增加饱和度
def HSV_saturation(x):
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            if x[i][j][1]*1.1>100:
                x[i][j][1]=100
            else:
                x[i][j][1]=x[i][j][1]*1.1
    return x

#HSV增加色调
def HSV_color(x):
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            if x[i][j][0]*1.1>360:
                x[i][j][0]=360
            else:
                x[i][j][0]=x[i][j][0]*1.05
    return x
```

```python
plt.figure(dpi=100)

plt.subplot(2,2,1)
plt.imshow(img_hsv)
plt.title('origin')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(HSV_bright(img_hsv))
plt.title('hsv_bright')
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(HSV_saturation(img_hsv))
plt.title('hsv_saturation')
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(HSV_color(img_hsv))
plt.title('hsv_color')
plt.axis('off')
```

##### 自定义

```python
#水平镜像
def flip(x):
    temp=x
    for i in range(x.shape[0]):
        for j in range(int(x.shape[1]/2)):
            temp1=temp[i][j+1]
            temp[i][j]=temp[i][x.shape[1]-j-1]
            temp[i][x.shape[1]-j-1]=temp1
    return temp
```

```python
temp2=img_rgb
plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.axis('off')
img_flip_h=flip(temp2)
plt.subplot(1,2,2)
plt.imshow(img_flip_h)
plt.title('水平翻转')
plt.axis('off')
```

#### 图像平移

```python
plt.figure(dpi=100)

plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title('origin')
plt.axis('off')

plt.subplot(1,2,2)
rows=img_rgb.shape[0]
cols=img_rgb.shape[1]
M = np.float32([[1,0,1000],[0,1,250]])
img_dst = cv2.warpAffine(img_rgb,M,(cols,rows))
plt.imshow(img_dst)
plt.title('distence')
plt.axis('off')
```

#### 图像旋转

```python
rows=img_rgb.shape[0]
cols=img_rgb.shape[1]
M1 = cv2.getRotationMatrix2D(((cols-1)/2.0,(rows-1)/2.0),25,1)
img_rotate1 = cv2.warpAffine(img_rgb,M1,(cols,rows))
M2 = np.float32([[np.sqrt(3)/2,0.5,0],[0.5,np.sqrt(3)/2,0]])
img_rotate2 = cv2.warpAffine(img_rgb,M2,(cols,rows))
plt.figure(dpi=100)
plt.subplot(1,2,1)
plt.title('rotate')
plt.imshow(img_rotate1)
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(img_rotate2)

plt.axis('off')
```

#### 透视变换

```python
pts1=np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2=np.float32([[0,0],[300,0],[0,300],[300,300]])
M=cv2.getPerspectiveTransform(pts1,pts2)
dst=cv2.warpPerspective(img_rgb,M,(1500,1000))

plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(dst)
plt.axis('off')
```
